```python


from pymoo.factory import get_visualization, get_reference_directions
from pymoo.factory import get_performance_indicator
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
from pandas.plotting import scatter_matrix
import pandas as pd
%matplotlib widget



M = 3
N = 5

rd = ref_dirs = get_reference_directions("energy", M, N, seed=1)
# print(ref_dirs)
hv = get_performance_indicator("hv", ref_point=np.array([1.001] * M))
print("Hypervolume:", round(hv.do(ref_dirs) * 100, 2), "%")
# get_visualization("scatter").add(ref_dirs).show()



fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
img = ax.scatter(rd[:,0], rd[:,1], rd[:,2])
# img = ax.scatter(rd[:,0], rd[:,1], rd[:,2], c=rd[:,3], cmap=plt.hot())
fig.colorbar(img)
ax.view_init(30,45)
plt.show()

# df = pd.DataFrame(rd, columns=['X', 'Y', 'Z', 'W'])
# scatter_matrix(df, alpha=0.2, figsize=(6, 6), diagonal='kde')
# sns.pairplot(df)
```
